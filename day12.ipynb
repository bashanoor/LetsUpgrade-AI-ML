# wilcoxson-sign Test & Freidman Test
this cannot be performed since there is no paired samples din this dataset

# Mann-Whitney

```python
from scipy.stats import mannwhitneyu
stat, p=mannwhitneyu(dataset.DistanceFromHome,dataset1.DistanceFromHome)
print(stat, p)
if p<0.05:
    print("Accept Null Hypothesis")
else:
    print("Reject Null Hypothesis")
```

# Chi Square

```python
from scipy.stats import chi2_contingency
gen=pd.crosstab(dataset1.Attrition,dataset1.Gender)
print(gen)
stats,p,dof,expected=chi2_contingency(gen)
print(stats, p)
if p<0.05:
    print("Accept Null Hypothesis")
else:
    print("Reject Null Hypothesis")
```

# One sample Test

```python
from scipy.stats import ttest_1samp
stats,p=ttest_1samp(dataset1.Age,37)
print(stats,p)
if p<0.05:
    print("Accept Null Hypothesis")
else:
    print("Reject Null Hypothesis")
```

# Tow Sample Separate T Test

```python
from scipy.stats import ttest_ind
stat, p=ttest_ind(dataset.DistanceFromHome,dataset1.DistanceFromHome)
print(stat, p)
if p<0.05:
    print("Accept Null Hypothesis")
else:
    print("Reject Null Hypothesis")
```
